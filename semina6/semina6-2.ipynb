## 7-3. 기하 정렬(Geometrical alignment)과 변환 추정

##### 지금까지는, 지역 정보만 사용한 매칭.

- 특징 벡터가 개별적으로 매칭 수행 -> 아웃라이어 매칭(거짓 긍정) 발생

> 아웃라이어를 찾아내고, 다시 정확한 점에 매칭시킬수 있을까

기하 정렬을 이용하여 인라이어 집합을 찾아내고 (global 정보를 활용), 변환 행렬을 추정한다.

<img src="https://raw.githubusercontent.com/bub3690/medical_lab_semina/main/semina6/resource/geo1.png" width="450" height="450">

변환행렬(T)를 구하면, 개구리 코나 발 처럼 영상에 없는 것의 위치도 추정이 되거나 아웃라이어를 제거하고 매칭이된다.

즉, 잡음,가림에도 강건하게 작동하는 알고리즘을 구현할 수 있다.

( T를 물체의 자세라고 하나??)

같은 물체가  같은 기하변환을 일으키는것을 <span style="color:blue">기하 정렬(Geometrical alignment)</span>이라 한다.

## 7-3-2. RANSAC

<img src="https://raw.githubusercontent.com/bub3690/medical_lab_semina/main/semina6/resource/ransac1.png" width="500" height="500">


에지 점들로 부터, 직선을 추정하는 방법이 RANSAC

대신, 입력이 대응점 쌍이고, 출력이 기하 변환 행렬(T)

> 초기 대응점도 랜덤으로 선택할까?

- PROSAC

RANSAC은 초기 대응점이 모두 인라이어야지 옳은답을 만들어냄.

< 매칭 점수(?)가 높을수록 선택 확률이 높은 방식에 따라, X에서 세 쌍을 선택 >

<img src="https://raw.githubusercontent.com/bub3690/medical_lab_semina/main/semina6/resource/ransac2.png" width="400" height="400">

---

적절한 반복수 K번을 추정하는 방법.

<img src="https://raw.githubusercontent.com/bub3690/medical_lab_semina/main/semina6/resource/ransac3.png" width="400" height="400">

q : 뽑은 대응점 쌍이 인라이어일 확률

q^3: 세 쌍모두 인라이어.

세쌍 모두가 인라이어야지 변환행렬 T가 될 확률이 크다.

즉, 확률 P(임계 확률)를 높이려면 K가 커져야한다. 계산시간을 더 줄수록 더 좋은 품질의 해를 찾는다.
